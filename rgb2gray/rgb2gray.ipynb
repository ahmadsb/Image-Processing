```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import signal as sg
import cv2
```

```python
I = plt.imread('shapes.jpg')
plt.imshow(I)
plt.colorbar()
print(I.shape)
height, width, channels = I.shape
```

```python
# provide permission to write on photo
I.setflags(write=1)
I.flags
```

```python
def rgb2Gray(img):
    for y in range(0,height):
        for x in range(0,width):
            img[y][x] =0.2126* img[y][x][0] + 0.7152 *img[y][x][1]+ 0.0722 *img[y][x][2]
    return img

I = rgb2Gray(I)
plt.imshow(I)
```
